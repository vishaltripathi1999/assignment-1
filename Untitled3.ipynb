```python

import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd

df = pd.read_csv('velocity_car.csv',header=None)
df.head()
```

```python
plt.scatter(df[0],df[1],color='b',alpha=0.9)
plt.title('ploting the graph of column 0 and column 1')
plt.xlabel('column 0')
plt.ylabel('column 1')
plt.show()

plt.scatter(df[1],df[2],color='hotpink',alpha=0.6)
plt.title('ploting the graph of column 1 and column 2')
plt.xlabel('column 1')
plt.ylabel('column 2')
plt.show()
```
